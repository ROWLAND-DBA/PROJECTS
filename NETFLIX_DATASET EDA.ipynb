```python
A boxplot of the "release year" shows that most movies on the Netflix network 
were released between 2002 and 2022. Also showing an outlier of release year 
to be well before 1940, with a consistent amount of releases between 
1960 and 2000. 
```

```python
sns.boxplot(x=net['release_year'])
```

```python
A scatter plot of duration by release year shows most of the movies released 
between 2000 and 2020 had longer durations.
```

```python
# Scatter plot of release_year vs title
plt.figure(figsize=(10, 6))
sns.scatterplot(x='duration', y='release_year', data=net)
plt.xlabel('duration')
plt.ylabel('release_year')
plt.title('Scatter Plot of release_year vs duration')
plt.show()
```

```python
A scatter plot of release year by type shows that the there was a consistent 
amount of movies being released between 1960 and 2020 as against the sparing 
releases of TV Shows up until 1980, which later picked up around 1985 to 2020 
```

```python
# Scatter Plot of release_year vs type
plt.figure(figsize=(10, 6))
sns.scatterplot(x='type', y='release_year', data=net)
plt.xlabel('type')
plt.ylabel('release_year')
plt.title('Scatter Plot of release_year vs type')
plt.show()
```

```python
net["type"].value_counts #(normalize = True)
```

```python
The count of the types(Movies and TV Shows) shows that MOvies are 69.6 percent 
of the movies on the network while TV Shows account for 30.3 percent of the 
shows on the network.
```

```python
net["type"].value_counts(normalize = True)
```

```python
An histogram plot of the count by type confirms the leaning that over 60% of 
content are Movies while over 30% are TV Shows. 
```

```python
plt.figure(figsize=(10,7))

sns.histplot(data=net, x=net["type"], bins=10)
plt.xticks(rotation=90)
plt.title('Histogram: count vs type')
          
plt.tight_layout()
plt.show()
```

```python
histogram shows Movies are about double the amount of TV Shows on the network
```

```python
net["type"].hist()
```

```python
A bar graph of count of the duration of the largest 20 movie or TV shows, shows 
that contents that are of one season have the largest count.
```

```python
net.duration.value_counts().nlargest(20).plot(kind='bar', figsize=(10,5))
```

```python
Plotting a histogram of count by duration shows that "season 1" contents have 
the highest count,implying that alot of their contents are abruptly ending in 
season 1 as against a continuation to more seasons.
```

```python
plt.figure(figsize=(12,7))

sns.histplot(data=net, x=net['duration'].head(10), bins=10)
plt.xticks(rotation=90)
plt.title('Histogram: duration by count')
          
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,7))

sns.histplot(data=net, x=net['duration'].head(50))
plt.xticks(rotation=90)
plt.title('Histogram: count of top 50 duration')
          
plt.tight_layout()
plt.show()
```

```python
A count based on their ratings shows that TV-MA (mature audience)has the highest
counts with 3207 counts, followed by TV-14 (not for 14 years and below) with a 
count of 2160 and TV-PG (parental guidance) having a count of 863.
Which draws the conclusion that there are more content for adults on the 
network.
```

# CONCLUSION

```python
Insights drawn from exploration of the dataset shows that 
1. The United States is the greatest contributor of contents to the network. 
2. Majority of the movies on the network were released between 2010 and 2020.
3. Movies account for more than half of the contents of the network as opposed 
   to TV-Shows.
4. Majority of the contents end in season one, which is not a plus for the 
   network.
5. A count of their contents ratings show that contents for Mature Audience has
   the highest count
6. Also established is that TV-MA has the highest mean value, G (general) 
    has the lowest lower quartile range and the lowest mean value. 
```
